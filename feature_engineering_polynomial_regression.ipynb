```python
import matplotlib.pyplot as plt
import numpy as np
import copy, math
```

```python
def compute_loss_value(x, y, w, b = 0):
    m = len(x)
    total_cost = 0
    for i in range(m):
        f_wb = w * x[i] + b
        total_cost += (f_wb - y[i]) ** 2 
    return total_cost

def compute_gradient(x, y, w, b):
    m = len(x)
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        err = (np.dot(x[i], w) + b) - y[i]
        dj_db += err
        dj_dw += err * x[i]
    dj_db /= m
    dj_dw /= m
    return dj_dw, dj_db

def gradient_descent(x, y, w_in, b_in, alpha, num_iters, gradient_function):
    w_out = w_in
    b_out = b_in
    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(x, y, w_out, b_out)
        w_out -= alpha * dj_dw
        b_out -= alpha * dj_db
            
    return w_out, b_out
```

```python

x = np.arange(0, 20, 1)
y = x**2 + 1
X = x.reshape(-1,1)

w_init = np.zeros(X.shape[1])
model_w, model_b = gradient_descent(X, y, w_init, 0, 0.006, 1000, compute_gradient)
print(model_w)
print(model_b)
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("no feature engineering")

plt.plot(x,X@model_w + model_b, label="Predicted Value")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python

x = np.arange(0, 20, 1)
y = x**2 + 1
X = x**2 
X = X.reshape(-1, 1) 
w_init = np.zeros(X.shape[1])
model_w, model_b = gradient_descent(X, y, w_init, 0, 0.0000701, 100000, compute_gradient)
print(model_w)
print(model_b)
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("Added x**2 feature")
plt.plot(x, np.dot(X,model_w) + model_b, label="Predicted Value"); plt.xlabel("x"); plt.ylabel("y"); plt.legend(); plt.show()
```
